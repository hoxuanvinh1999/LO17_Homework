## Tokenization

Tout d'abord, nous devons charger un fichier txt. Pour simplifier, j'utilise un fichier nommé test.txt comme exemple et le place dans le même dossier que ce fichier.

```python
import nltk
import os
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import word_tokenize, pos_tag, ne_chunk
from collections import Counter
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import stopwords
textfile = open('test.txt','r')
data = textfile.read()
textfile.close
print(data)
```

### Découpage en phrases

```python
seg_sentences = sent_tokenize(data)
print(seg_sentences)
```

### Découpage en mots

```python
seg_word = word_tokenize(data)
print(seg_word)
```

## Part-Of-Speech Tagging (POS Tagging)

```python
pos_tagging = nltk.pos_tag(seg_word, tagset="universal")
for couple in pos_tagging:
    print (couple)
```

## Lemmatization


```python
from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
wordnet_map = {
    "N": wordnet.NOUN,
    "V": wordnet.VERB,
    "J": wordnet.ADJ,
    "R": wordnet.ADV
}
pos_tagged_text = nltk.pos_tag(seg_word, tagset="universal")
# map the pos tagging output with wordnet output
pos_tagged_text = [
        (word, wordnet_map.get(pos_tag[0])) if pos_tag[0] in wordnet_map.keys()
        else (word, wordnet.NOUN)
        for (word, pos_tag) in pos_tagged_text
]
data_lemmatization = []
for couple in pos_tagged_text:
    data_lemmatization.append(lemmatizer.lemmatize(couple[0],pos = couple[1]))
for before,after in zip(seg_word,data_lemmatization):
    print(before, '-->' ,after )
```

## Reconnaissance d’entités nommées


```python
pos_tags = pos_tag(seg_word)
named_entities = ne_chunk(pos_tags)
print(named_entities)
```
