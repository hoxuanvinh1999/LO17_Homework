2. Ecrire une fonction qui prend un synset en paramètre et qui renvoie la liste des mots
contenus dans le synset.

```python
def list_word_synset(name):
    root_synset = wn.synset(name)
    result = list(set([str.lower(w) for s in root_synset.closure(lambda s:s.hyponyms()) for w in s.lemma_names() ]))
    return(result)

print(list_word_synset('car.n.01'))
```
